```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("iris")
sns.histplot(data=data, x="sepal_length", kde=True)
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")

plt.show()

```
